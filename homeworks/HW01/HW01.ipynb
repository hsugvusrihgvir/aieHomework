1-2 Примеры создания массивов

```python
import numpy as np

lst = [1, 2, 6, 3 , 9586, 'urhw']
arr1 = np.array(lst)
print("массив из списка 1: ", arr1)

lst2 = [1, 2, 6, 3]
arr12 = np.array(lst2)
print("массив из списка 2: ", arr12)

# zeros (заполняет нулями)
arr2 = np.zeros(10)
print("массив из нулей: ", arr2)

# ones (заполняет единицами)
arr3 = np.ones((6, 3))
print("Массив из единиц:")
print( arr3)

# arange
arr4 = np.arange(0, 100, 11)
print("Массив с помощью arange: ", arr4)

# другие
print()
arr5 = np.full((5, 3), 9) # заполняет числом
print(arr5)
print()
arr6 = np.zeros_like(arr3) # такой же формы
print(arr6)
```

4 Примеры индексирования и срезов

```python
arr = np.arange(0, 400).reshape((20, 20))

el = arr[3, 3]
print("получение отдельного элемента:", el)

sr = arr[4:9, 8:15]
print("извлечение подмассива с помощью среза:", sr)
print(sr.shape)
```

5 Примеры векторных операций и базовых статистик

```python
a = np.array([67, 23, 89, 54])
b = np.arange(56, 100, 14)

# поэлементное сложение
print("a + b =", a + b)
# умножение массивов
print("a * b =", a * b)

print("Средняя величина:", a.mean())
print("Сумма:", a.sum())
print("Минимальное:", a.min())
print("Максимальное:", a.max())
```
